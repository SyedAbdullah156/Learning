Pandas provides two types of classes for handling data:

- **Series**: a one-dimensional labeled array holding data of any type such as integers, strings, Python objects etc.  
- **DataFrame**: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

````python
import numpy as np
import pandas as pd
import seaborn as sns
````

# Series
Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index. 

````python
# General Syntax for Creating Series 
# s = pd.Series(data, index=index)

s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
print(s)
print(s.index)
print()

s = pd.Series(np.random.randn(5))
print(s)
print()

s = pd.Series({'a': 1, 'b': 2, 'c': 3})
print(s)
print()

s = pd.Series({'a': 1.0, 'b': 2.0, 'c': 3.0}, index=['a', 'b', 'c', 'd'])
print(s)
print()

s = pd.Series(5.0)
print(s)
print()

s = pd.Series([5.0, 5.0])
print(s)
print()

s = pd.Series(5.0, index=["a", "b", "c", "d", "e"])
print(s)
print()
````

Series acts very similarly to a ndarray and is a valid argument to most NumPy functions

````python
# ndarray and Series Comparison

s = pd.Series([5.0] * 5)
print(s)

print(np.exp(s))
print(s.dtype)
print(s.array)
print(s.to_numpy().__array__) # converting the pandas Series to numpy ndarray
````

````python
# Vectorized Operations

print(s + s)
print(s * 2)
print(s @ s)
print(s * s)
print(np.exp(s))
````

## Indexing in pandas

Pandas supports all the types (borrowed from NumPy) plus label-based indexing.

### 1. Label-based indexing (.loc)

Uses labels (index names, column names). 👉 Think of .loc as “by name”.

Inclusive of endpoints for slicing (different from Python/NumPy).

````python
s = pd.Series([10, 20, 30, 40, 50], index=["a", "b", "c", "d", "e"])
print(s)

# Single Labels
print(s.loc["a"])
print(s.loc["d"])

# Multiple Labels (Fancy Indexing)
print(s.loc[["a","c","e"]])

# Slicing with endpoint Included (Differ from Numpy or Python Slicing)
print(s.loc["b":"d"])

# Creating Mask that returns a pd.Series as result with key(index): bool(True or False)
print(s > 25)
print(s.loc[s > 25])
print(s.loc[[True, True, False, False, False]])
````

### 2. Position-based indexing (.iloc)

Uses the integer position (like NumPy).

````python
s = pd.Series([10, 20, 30, 40, 50], index=['a','b','c','d','e'])
print(f"s\n",s)
# Single Location
print(f"s.iloc[2]\n",s.iloc[2])           

# Multiple Locations (Fancy Indexing)
print(f"s.iloc[[0,4]]\n",s.iloc[[0,4]])      
print(f"s.iloc[0,4]: Error\n")      
# Slicing
print(f"s.iloc[1:4]\n",s.iloc[1:4])         
print(f"s.iloc[1:]\n",s.iloc[1:])
print(f"s.iloc[:-1]\n",s.iloc[:-1])
print(f"s.iloc[1:] + s.iloc[:-1]\n",s.iloc[1:] + s.iloc[:-1])
print(f"s.iloc[::2]\n",s.iloc[::2])

# Masking (Here, the mask is position-based — it must match the length of the Series.)
mask = [True, False, True, False, True]
print("s.iloc[mask]\n", s.iloc[mask])
````

### 3. Simple Indexing
Same as Label base (loc) indexing

````python
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# Using keys indexing without .loc
print(s['a'])   
print(s['b']) 

# Using keys slicing without .loc
print(s['a':'c'])

# Using integer indexing witout .iloc
print(s[1])

# Using integer slicing without .iloc
print(s[0:3])

# Error in case of int indexing without .iloc
print(s[0])
````

# DataFrame

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object. Like Series, DataFrame accepts many different kinds of input:


````python
# Making dictionary from Series 
dict1 = {
    "one": pd.Series([1.0,2.0,3.0], index=["a","b","c"]), # (index1's length == array1's length)
    "two": pd.Series([1.0,2.0,3.0,4.0], index=["a","b","c","d"]) # (index2's length == array2's length)
}

# Making DataFrame from dictionary of Series (can be of different lengths)
df = pd.DataFrame(dict1)
print(df)

print(df.index)
print(df.columns)
print()

# Making dictionary from Lists
dict2 = {
    "Name": ["Ali", "Sara", "John"],
    "Age": [22, 25, 29],
    "City": ["Lahore", "Karachi", "New York"]
}

# Making DataFrame from dictionary of Lists with default indexes
df = pd.DataFrame(dict2)
print(df)

print(df.index)
print(df.columns)
print()

# Making DataFrame from dictionary of Lists with custom indexes
df = pd.DataFrame(dict2, index=['a','b','c'])
print(df)

print(df.index)
print(df.columns)
print()

# Making DataFrame from subset of dictionary (selecting only columns we need)
s1 = pd.Series([1,2,3], index=['a','b','c'])
s2 = pd.Series([4,5],   index=['a','b'])

df = pd.DataFrame({"A": s1, "B": s2}, index=['b','c'])
print(df)
````

**Remember that if we pass index array individually to each of the column/ field/ key in the dataframe then that index array must have length == length of values in that field. But when we pass another field in the DataFrame with its own index array the length should be same to the field2's values but the index1 and index2's length both need not to be same. But when we pass the index array separately in the end after making the dictionary with only list of values no index for each Series separately in that case the index array should be of same size as of both arrays**

````python
# Making dictionary from ndarray
# All arrays must be of same length
dict = {
    "one": [1.0, 2.0, 3.0, 4.0],
    "two": [4.0, 3.0, 2.0, 1.0]
}

# Making DataFrame with no index information (set 0 to N by default)
df = pd.DataFrame(dict)
print(df)

# Making DataFrame with index information (index array length must be same as of all arrays)
df = pd.DataFrame(dict, index=["a","b","c","d"])
print(df)

# Making DataFrame with index and columns defined
df = pd.DataFrame(np.random.rand(3, 2), columns=['foo', 'bar'], index=['a', 'b', 'c']) # np.random.random((2,3)) it expects a tuple as argument, While np.random.rand(d0, d1, ..., dm) expects individual dimensions
print(df)
````

````python
# Making DataFrame from list of list

df = pd.DataFrame(
    [[1, 2, 3], [4, 5, 6], [7, 8, 9]],
    columns=["A", "B", "C"]
)
print(df)

# Making DataFrame from list of dictionaries 
df = pd.DataFrame(
    [{"a": 1, "b": 2},
     {"b": 5, "c": 10}]
)
print(df)

df = pd.DataFrame({'a': i, 'b': i * 2} for i in range(3))
print(df)
````

````python
# Making Complex DataFrame with conditions on dtypes of columns

''' 
np.zeros((2,), ...) → makes an array of 2 elements, initialized to zero.

dtype=[("A", "i4"), ("B", "f4"), ("C", "a10")] → defines a structured dtype:
"A" → field name, type: i4 = 4-byte int (int32)
"B" → field name, type: f4 = 4-byte float (float32)
"C" → field name, type: S10 = string of length 10 
'''

data = np.zeros((2,), dtype=[("A", "i4"), ("B", "f4"), ("C", "S10")])

# NumPy matches tuple elements positionally with fields (A, B, C).
data[:] = [(1, 2.0, "Hello"), (2, 3.0, "World")]

print(data)

# Making DataFrame with Default indexes
df = pd.DataFrame(data)
print(df)

# Making DataFrame with custom indexes (size = no of rows)
df = pd.DataFrame(data, index=["first", "second"])
print(df)

# Making DataFrame with changed order of columns
df = pd.DataFrame(data, columns=["C", "A", "B", "D"])
print(df)
````

````python
# Making DataFrame from 2D np.array
df = pd.DataFrame(np.random.randn(6, 4), index=list("012345"), columns=list("ABCD"))
print(df)
print()

# df["name"] → looks up column "name" not rows index
print(df["A"])
print()
print(df["B"])
print()
# print(df["0"]) # error no column with name "0"

# df.loc["name"] → looks up row with label "name"
print(df.loc["0"])
print()
print(df.loc["1"])
print()
# print(df.loc["A"]) # error as no row with index = "A" 
````

## 1️⃣ Plain Brackets `df[...]`

* **`df['col_name']`** → Select **a single column** (returns a Series).
* **`df[['col1','col2']]`** → Select **multiple columns** (returns a DataFrame).
* **`df[boolean_mask]`** → Select **rows where mask is True** (boolean indexing).
* **`df[start:stop]`** → Slice **rows by position** (like `iloc` slice, only rows).

**Examples:**

```python
df['age']          # Series of 'age'
df[['name','age']] # DataFrame of two columns
df[df['age'] > 18] # Rows where age>18
df[2:5]            # Rows 2,3,4 (position-based slice)
```

> ⚠️ Cannot select rows AND columns at the same time with plain brackets.

---

## 2️⃣ `.iloc[]` → **Integer-location based**

* **Rows and columns are indexed by integer positions only.**
* Syntax: `df.iloc[rows, cols]`

  * `rows` → integer, list of integers, slice
  * `cols` → integer, list of integers, slice

**Examples:**

```python
df.iloc[0]        # Row 0 as Series
df.iloc[0:3]      # Rows 0,1,2 as DataFrame
df.iloc[:,0]      # Column 0 (all rows)
df.iloc[0,0]      # Row 0, Col 0 scalar
df.iloc[[0,2],[1,3]]  # Rows 0,2 and Columns 1,3
```

---

## 3️⃣ `.loc[]` → **Label-based**

* **Rows and columns indexed by labels (row index / column names).**
* **`df.loc[boolean_mask]`** → Select **rows where mask is True** (boolean indexing).
* Syntax: `df.loc[rows, cols]`

  * `rows` → label, list of labels, slice of labels, boolean mask
  * `cols` → column name(s), list, slice

**Examples:**

```python
df.loc['a']        # Row with index label a (can differ from position!)
df.loc[:, 'age'] # Column 'age', all rows
df.loc[['a','b'],'age']  # Scalar at index a and b, column 'age'
df.loc[['a':'c'], ['name','age']] # Rows a, b, c and columns 'name','age'
df.loc[df['age']>18, 'name']   # Conditional selection of column
```

---

## ✅ Summary Table

| Access         | Row                        | Column              | Notes                                       |
| -------------- | -------------------------- | ------------------- | ------------------------------------------- |
| `df[...]`      | slice by pos, boolean mask | single/multiple col | Cannot mix row & col explicitly             |
| `df.iloc[...]` | integer pos                | integer pos         | Position-only, always numeric               |
| `df.loc[...]`  | label / boolean mask       | label(s)            | Label-based, exact match or slice by labels |

---

### Examples for all combinations:

```python
# All rows, single column
df['age']           # plain
df.iloc[:,1]        # iloc
df.loc[:, 'age']    # loc

# Single row, all columns
df.iloc[0,:]        # iloc by position
df.loc[0,:]         # loc by label
# df[...] cannot do this directly

# Slice rows and specific columns
df.iloc[0:3, [0,2]]
df.loc[0:3, ['name','age']]
```

### 1.Column Selection
Simple indexing with []. Always selects columns by label (never rows).
df[something] without .loc/.iloc expects a column label (string, list of strings, boolean mask).

````python
print(df)
print()

# Filtering Columns

print(df['V'])                                      # DataFrame column V
print()

print(df[['V','X']])                                # DataFrame with columns V and X
print()

# df['a']                                           # KeyError because 'a' is a ROW label, not a column

print(df.V)                                           # DataFrame with colum V (works for valid)
print()

# Filtering Rows and Columns
print(df['V'].loc['a'])                             # df['V'] returns pd.Series which can be indexed using label 'a'
print(df['V']['a'])                                 # df['V'] returns pd.Series which can be indexed using label 'a'
print(df['V'].iloc[0])                              # df['V'] returns pd.Series which can be indexed using index 0
print(df['V'][0])                                   # deprecated 
# print(df['a']['V'])                               # error as 'a' is not a column
print()
````

### 2.Row selection by label
Label indexing with ".loc".

````python
# Syntax: df.iloc[RowStartIncluded : RowEndIncluded : RowStep, ColumnStartIncluded : ColumnEndIncluded : ColumnStep]  

print(df)
print()

# Filtering Rows based on labels

print(df.loc['a'])                              # Row 'a' (Series)    
print()   

print(df.loc[['a','c']])                        # Rows 'a' and 'c' (DataFrame)
print()

print(df.loc['a':'c'])                          # Slice rows by label (inclusive: a and b)
print()

# Filtering Rows and Columns based on labels

print(df.loc['a', ['V','X']])                   # Row 'a', only columns V and X
print(df.loc['a'].loc[['V','X']])               # df.loc['a'] returns a pd.Series that can be indexed using column 'V'
print()

print(df.loc[['b', 'c'], 'V'])                  # Rows 'b' and 'c', only column V
print(df.loc[['b','c']]['V'])                   # df.loc[['b','c']] returns sub-DataFrame with rows = b, c and all columns. On that DataFrame we futher filter it using column 'V' simple indexing.
print(df.loc[['b','c']].iloc[:, 0])             # df.loc[['b','c']] returns sub-DataFrame with rows = b, c and all columns. On that DataFrame we futther filter it using slicing index on position of rows and columns. Such that all rows are included (b, c) but only the first column is included (V).
# print(df.loc[['b', 'c']].loc['V'])            # error because df.loc[['b','c']] returns sub-DataFrame with rows = b,c and all columns. On that DataFrame loc accepts only b, c as arg (rows) not columns.
print()

print(df.loc['a', 'V':'Z':2])
print()

print(df.loc['a'::2, 'V'::2])
print()

print(df.loc['a': 'c' : 2, 'V': 'Z': 2])
print()

print(df['V'])
print(df['V'] == 10)                             # df['V'] is column selection it returns a series which is then compared with 10. It transforms from [0,5,10,15] to [False,False,True,False].
print(df.loc[df['V'] == 10])                     # Based on the mask generated when this is passed to .loc[rows, column] the rows are filtered based on it. So this prints any row that contains the value of column 'V' as 10
print(df.loc[df['V'] == 10, 'X'])                # This prints the value of column X based on filtered row.
````

### 3. Row Selection by Position
Position indexing with ".iloc". Always selects rows and columns by indexes (0 to N).

````python
# Syntax: df.iloc[RowStartIncluded : RowEndExcluded : RowStep, ColumnStartIncluded : ColumnEndExcluded : ColumnStep]

print(df)
print()

# Filtering Only Rows

print(df.iloc[0])                       # Only row 0 or 'a'
print()

# print(df[0])                          # error as there is no column with 0
# print()

print(df.iloc[[0, 3]])                  # Only row 0 and 2 / 'a' and 'c'
print()

print(df.iloc[0:3])                     # Rows from 0 to 3 (0,1,2)
print()

print(df[0:3])                          # deprecated although still works 
print()

# Filtering Rows with columns

print(df.iloc[:, [0, 2]])               # All rows but columns 'A' and 'C'
print()

print(df.iloc[0::2, 1::2])
print()

print(df.iloc[[0, 2], [1, 3]])
print()
````

# Filling Values

````python
df = sns.load_dataset('titanic')
print(df[:3])
print(df.isna()[:3]) # same as df.isnull()[:3]
````

````python
print(df['deck'][:5].tolist())
print(df['deck'][:5].isnull().tolist()) # same as df.isna().tolist()
````

````python
# Count missing values per column
missing_counts = df.isnull().sum()
print(missing_counts)
````

````python
# Percentage of missing values per column
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
````

````python
df = pd.DataFrame({
    "name": ["Ali", "Sara", "John", None],
    "age": [25, None, 30, 22],
    "city": ["Lahore", "Karachi", None, "Islamabad"] 
})
print(df)

# Method-1: Fill all missing values with one value
print(df.fillna("Unknown"))

# Method-2: Fill per column with a dictionary
print(df.fillna({"name": "No Name", "age": 0, "city": "No City"}))

# Method-3: Fill with statistical values (common practice)
print(df.fillna({
    "name": df["name"].mode()[0],
    "age": df["age"].median(),
    "city": df["city"].mode()[0]
}))

# Method-4: Fill only one column
print(df["age"].fillna(df["age"].median()))
print(df)

# But these methods does not modify the original df. For that modification we have to pass inplace=True
df.fillna({
    "name": df["name"].mode()[0],
    "age": df["age"].median(),
    "city": df["city"].mode()[0]
}, inplace=True)

# or 
# df["age"] = df["age"].fillna(df["age"].mean())
# df["name"] = df["name"].fillna(df["name"].mode()[0])

print(df)
````

# ISNULL and NOTNULL

````python
print(df[df["age"].isnull()])    # Select rows where age is missing
print()

print(df[df["city"].notnull()])  # Select rows where city is present
````

## Type conversions

````python
# Sometime we need type conversions like in this case of one-hot encoding
np.random.seed(42)
n_train = 10
grades = ['A', 'B', 'C', 'D', 'E']

y_train = pd.Series(np.random.choice(grades, n_train))

print(y_train[:5])
print()

# Vectorized comparison → produces a boolean array, then convert to int
# Convert labels to a DataFrame of one-hot style
y_train_ova = pd.DataFrame({
    'A': (y_train == 'A').astype(int),
    'B': (y_train == 'B').astype(int),
    'C': (y_train == 'C').astype(int),
    'D': (y_train == 'D').astype(int),
    'E': (y_train == 'E').astype(int)
})

print(y_train_ova.head())
````
