```python
%%manim -qm Function1

class Function1(Scene):
    def construct(self):
        plot_axes = Axes(
            # Changed the step size to 200 so only 5 labels generate per axis
            x_range=[0, 1000, 200],
            y_range=[0, 1000, 200],
            x_length=9,
            y_length=5.5,
            axis_config={
                "font_size": 20,
                "include_numbers": True,
            },
            tips=False
        )

        y_label = plot_axes.get_y_axis_label(
            "y",
            edge=LEFT,
            direction=LEFT,
            buff=0.4
        )
        x_label = plot_axes.get_x_axis_label(
            "x",
            edge=DOWN,
            direction=DOWN,
            buff=0.4
        )
        plot_labels = VGroup(x_label, y_label)

        # Plotting the function across the new large domain
        graph = plot_axes.plot(
            lambda x: np.sqrt(x**2 - 3*x),
            x_range=[4, 1000], 
            color=BLUE
        )

        self.play(
            Create(plot_axes),
            Write(plot_labels)
        )
        self.wait(0.5)
        self.play(
            Create(graph),
            run_time=2
        )
        self.wait(2)
```

```python
%%manim -qm Function1

class Function1(Scene):
    def construct(self):
        plot_axes = Axes(
            x_range=[-5, 5, 1],
            y_range=[-5, 5, 1],
            x_length=9,
            y_length=5.5,
            axis_config={
                "font_size": 20,
                "include_numbers": True,
            },
            tips=False
        )

        y_label = plot_axes.get_y_axis_label("y")
        x_label = plot_axes.get_x_axis_label("x")
        plot_labels = VGroup(x_label, y_label)

        # Define domains that avoid absolute zero to prevent vertical line artifacts
        left_domain = [-5, -0.1]
        right_domain = [0.1, 5]
        
        graphs = VGroup()
        
        # Plotting each function in two halves (left side and right side of the asymptote)
        # 1/x (BLUE)
        graphs += plot_axes.plot(lambda x: 1/x, x_range=left_domain, color=BLUE)
        graphs += plot_axes.plot(lambda x: 1/x, x_range=right_domain, color=BLUE)
        
        # 4/x (GREEN)
        graphs += plot_axes.plot(lambda x: 4/x, x_range=left_domain, color=GREEN)
        graphs += plot_axes.plot(lambda x: 4/x, x_range=right_domain, color=GREEN)
        
        # 1 + 4/x (RED)
        graphs += plot_axes.plot(lambda x: 1 + 4/x, x_range=left_domain, color=RED)
        graphs += plot_axes.plot(lambda x: 1 + 4/x, x_range=right_domain, color=RED)

        self.play(
            Create(plot_axes),
            Write(plot_labels)
        )
        self.wait(0.5)
        self.play(
            Create(graphs),
            run_time=2
        )
        self.wait(2)
```

```python
%%manim -qm Function1

class Function1(Scene):
    def construct(self):
        # 1. Update axes: 
        # y_range is now [-50, 50, 10] to step by 10.
        plot_axes = Axes(
            x_range=[-5, 5, 1],
            y_range=[-50, 50, 10], 
            x_length=9,
            y_length=5.5,
            axis_config={
                "include_numbers": True,
                "font_size": 24,
            },
            tips=False,
        )

        y_label = plot_axes.get_y_axis_label("y")
        x_label = plot_axes.get_x_axis_label("x")
        plot_labels = VGroup(x_label, y_label)

        # 2. Add the Bounding Box
        # Using coords_to_point to map graph coordinates to screen coordinates
        # The rect spans from x=-5 to 5, and y=0 to 50
        bottom_left = plot_axes.coords_to_point(-5, 0)
        top_right = plot_axes.coords_to_point(5, 50)
        
        # Calculate width/height based on coordinates
        box_width = top_right[0] - bottom_left[0]
        box_height = top_right[1] - bottom_left[1]
        
        bounding_box = Rectangle(
            width=box_width, 
            height=box_height,
            stroke_color=YELLOW
        ).move_to(plot_axes.coords_to_point(0, 25)) # Center between 0 and 50

        c = 2
        graphs = VGroup()
        graphs += plot_axes.plot(lambda x: x**2, color=RED)
        graphs += plot_axes.plot(lambda x: c * x**2, color=GREEN)
        graphs += plot_axes.plot(lambda x: 1/c * x**2, color=BLUE)

        # Animation sequence
        self.play(Create(plot_axes), Write(plot_labels))
        self.play(Create(bounding_box)) # Added bounding box animation
        self.wait(0.5)
        self.play(Create(graphs), run_time=2)
        self.wait(2)
```

```python
class LogScalingExample(Scene):
    def construct(self):
        ax = Axes(
            x_range=[0, 10, 1],
            y_range=[-2, 6, 1],
            tips=False,
            axis_config={"include_numbers": True},
            y_axis_config={"scaling": LogBase(custom_labels=True)},
        )

        # x_min must be > 0 because log is undefined at 0.
        graph = ax.plot(lambda x: x ** 2, x_range=[0.001, 10], use_smoothing=False)
        self.add(ax, graph)



# don't remove below command for run button to work
%manim -qm -v WARNING LogScalingExample
```
